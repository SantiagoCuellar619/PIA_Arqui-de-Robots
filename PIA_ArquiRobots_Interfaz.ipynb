```python
#Programación para interfaz de control del brazo robótico SCARA
from tkinter import *
from pyArduino import *

def onClosing():
    arduino.sendData([0,0,0])
    arduino.close()
    print("Arduino desconectado")
    root.destroy()

def getAngle(int):
    q1Angle.set(q1Scale.get())
    q2Angle.set(q1Scale.get())
    zAngle.set(q1Scale.get())

    status.set("Modo manual")

def q1MinusOk():
    q1Angle.set(int(q1Angle.get()) - int(q1Value.get()))
    q1Scale.set(q1Angle.get())

def q1PlusOk():
    q1Angle.set(int(q1Angle.get()) + int(q1Value.get()))
    q1Scale.set(q1Angle.get())

def q2MinusOk():
    q2Angle.set(int(q2Angle.get()) - int(q2Value.get()))
    q2Scale.set(q2Angle.get())

def q2PlusOk():
    q2Angle.set(int(q2Angle.get()) + int(q2Value.get()))
    q2Scale.set(q2Angle.get())

def zMinusOk():
    zAngle.set(int(zAngle.get()) - int(zValue.get()))
    zScale.set(zAngle.get())

def zPlusOk():
    zAngle.set(int(zAngle.get()) + int(zValue.get()))
    zScale.set(zAngle.get())

def enviar():
     q1Angle.set(q1Value.get())
     q2Angle.set(q2Value.get())
     zAngle.set(zValue.get())

     q1Scale.set(q1Angle.get())
     q2Scale.set(q2Angle.get())
     zScale.set(zAngle.get())

def home():
     q1Angle.set(0)
     q2Angle.set(0)
     zAngle.set(0)

     q1Scale.set(q1Angle.get())
     q2Scale.set(q2Angle.get())
     zScale.set(zAngle.get())

     global isRun
     isRun = False

def guardar():

     global q1
     global q2
     global z
     q1.append(q1Angle.get())
     q2.append(q2Angle.get())
     z.append(zAngle.get())

     status.set("Posiciones guardadas : " + str(q1Angle.get()) + "," + str(q2Angle.get()) + "," + str(zAngle.get()))

def ejecutar():
     global isRun
     isRun = True
     status.set("Modo automático")

def borrar():
     global isRun
     isRun = False

     q1.clear()
     q2.clear()
     z.clear()

     home()

     status.set("Posiciones borradas")

def callback():
    global k

    if isRun:
        arduino.sendData([q1[k],q2[k],z[k]])
        if k < len(q1) -1:
           k = k + 1
           else:
             k = 0
    else:
        arduino.sendData([q1Angle.get(),q2Angle.get(),zAngle.get()])


    root.after(1000,callback)  #Llamar a callback despues de 1 seg

# Comunicación serial
port = "COM6"
arduino = serialArduino(port)


# Interfaz de usuario
root = Tk()
root.protocol("800x600")
root.geometry("WM_DELETE_WINDOW",onClosing)
root.title("Interfaz de control")

title = Label(root,text="Control Robot SCARA",font=("Arial",18,"bold"))
title.place(x = 250, y = 10)

# Movimiento Eslabon 1
q1Angle = IntVar()

q1Scale = Scale(root, command = getAngle, label = "Eslabon 1", font=("Arial 14"), from_ = -90,to = 90, orient = HORIZONTAL, length= 300)
q1Scale.place(x = 40, y = 50)

q1Minus = Button(root, text = "-°G1", font=("Arial 14"), command = q1MinusOk, height = 1, width= 6)
q1Minus.place(x = 40, y = 140)

q1Value = Entry(root,width = 6, font=("Arial 18"))
q1Value.place(x = 140, y = 142)
q1Value.insert(0,"1")

q1Plus = Button(root, text = "+°G1", font=("Arial 14"), command = q1PlusOk, height = 1, width= 6)
q1Plus.place(x = 240, y = 140)

# Movimiento Eslabon 2
q2Angle = IntVar()

q2Scale = Scale(root, command = getAngle, label = "Eslabon 2", font=("Arial 14"), from_ = -90,to = 90, orient = HORIZONTAL, length= 300)
q2Scale.place(x = 40, y = 200)

q2Minus = Button(root, text = "-G2", font=("Arial 14"), command = q2MinusOk, height = 1, width= 6)
q2Minus.place(x = 40, y = 290)

q2Value = Entry(root,width = 6, font=("Arial 18"))
q2Value.place(x = 140, y = 292)
q2Value.insert(0,"1")

q2Plus = Button(root, text = "+G2", font=("Arial 14"), command = q2PlusOk, height = 1, width= 6)
q2Plus.place(x = 240, y = 290)

# Movimiento Eslabon z
zAngle = IntVar()

zScale = Scale(root, command = getAngle, label = "Eje Z", font=("Arial 14"), from_ = -50,to = 50, orient = HORIZONTAL, length= 300)
zScale.place(x = 40, y = 350)

zMinus = Button(root, text = "- z ", font=("Arial 14"), command = zMinusOk, height = 1, width= 6)
zMinus.place(x = 40, y = 440)

zValue = Entry(root,width = 6, font=("Arial 18"))
zValue.place(x = 140, y = 442)
zValue.insert(0,"1")

zPlus = Button(root, text = "+ z", font=("Arial 14"), command = zPlusOk, height = 1, width= 6)
zPlus.place(x = 240, y = 440)

# Trabajo manual/automatico
status = StringVar(root,"Modo Manual")
labelU= Label(root, textvariable = status)
labelU.place( x = 530, y = 150)

# Enviar programa
sendBtn = Button(root,text="Enviar",font=("Arial 14"), command = enviar)
sendBtn.place(x = 40, y = 520)

# Regresar a casa
homeBtn = Button(root,text="Casa",font=("Arial 14"), command = casa)
homeBtn.place(x = 140, y = 520)

# Guardar programa
q1 = []
q2 = []
z = []
saveBtn = Button(root,text="Guardar",font=("Arial 14"), command = guardar)
saveBtn.place(x = 480, y = 200)

# Ejecutar programa
isRun = False
k = 0
runBtn = Button(root,text="Ejecutar",font=("Arial 14"), command = ejecutar)
runBtn.place(x = 620, y = 200)

# Borrar datos
deleteBtn = Button(root,text="Borrar",font=("Arial 14"), command = borrar)
deleteBtn.place(x = 480, y = 250)



root.after(500,callback)  #Llamar a callback despues de 500ms
root.mainloop()
```
